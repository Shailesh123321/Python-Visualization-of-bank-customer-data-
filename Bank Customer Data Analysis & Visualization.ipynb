```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('bank_transactions.csv')
df = data
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
from datetime import datetime
```

```python
current_year = 2023
df['CustomerDOB'] = pd.to_datetime(df['CustomerDOB'])
df['Age'] = current_year - df['CustomerDOB'].dt.year
df['FormattedBirthDate'] = df['CustomerDOB'].dt.strftime('%m/%d/%Y')
```

```python
df
```

```python
df = df[df['Age'].between(0,100)]
df
```

```python
df['sex'] = df['CustGender'].replace(['F', 'M','T'],[0,1,2])
df
```

```python
top_100_transactions = df.sort_values(by = 'TransactionAmount (INR)', ascending = False )
```

```python
newdata = top_100_transactions.head(100)
newdata
```

## Transaction VS Age

```python
plt.scatter(newdata['Age'], newdata['TransactionAmount (INR)'])
plt.xlabel('Age')
plt.ylabel('Amount')
plt.title('Top 100 Transaction by Age')
plt.show()
```

## Cities with Highest Transactions

```python
group_cities = df.groupby('CustLocation')['TransactionAmount (INR)'].sum().reset_index()
top10cities =group_cities.sort_values(by = 'TransactionAmount (INR)', ascending = False).head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top10cities['CustLocation'], top10cities['TransactionAmount (INR)'])
plt.xlabel('Cities')
plt.ylabel('Amount')
plt.title('Top 10 Transaction by cities')
plt.xticks(rotation=45)
plt.show()
```

```python
top10cities
```

## Citywise transactions with M/F

```python
group_cities = df.groupby(['CustLocation', 'CustGender'])['TransactionAmount (INR)'].sum().reset_index()
new_cities_data = group_cities.sort_values(by='TransactionAmount (INR)', ascending= False).head(20)
new_cities_data
```

```python
plt.figure(figsize= (12,6))

sns.barplot(x='CustLocation', y='TransactionAmount (INR)', hue = 'CustGender', data= new_cities_data)
plt.xlabel('Cities')
plt.ylabel('Amount')
plt.title('Transaction_Amount_Gender/Citywise')
plt.xticks(rotation=45)
plt.show()
```

## Comparison of % of transactions between male and female

```python
total_transactions = df.groupby('CustGender')['TransactionAmount (INR)'].sum()
```

```python
total_transactions_all = df['TransactionAmount (INR)'].sum()
```

```python
percentage_transactions = (total_transactions/total_transactions_all)*100
print(percentage_transactions)
```

```python
plt.figure(figsize=(6,6))
plt.pie(total_transactions, labels=total_transactions.index, autopct='%1.1f%%', startangle = 125)
plt.title('Percentage of Transactions by Gender')
plt.show()
```
